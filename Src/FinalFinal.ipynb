```python
import networkx as nx
import collections
import matplotlib.pyplot as plt
import numpy as np

file = "../data/twitter_sentiment_200.gexf"     #CHANGE to your file
G = nx.read_gexf(file)
#print(yoyo.nodes['Shehla_Rashid'])
#print(yoyo.nodes['Shehla_Rashid']['sent'])
```

```python
v10 = np.load('heh10.npy')
v25 = np.load('heh10.npy')
heh1 = G.subgraph(v10)
heh2 = G.subgraph(v25)
```

```python
#CHANGE PLOT TITLES ALONG WITH FILENAMES IN PLT.SAVEFIG

def compare_degree(s1,s2):
    a = [i[1] for i in s1.degree()]
    a2 = collections.Counter(a)
    a2v = list(a2.values())
    print(a2v)
    plt.bar(a2.keys(), np.log(a2v), width = 1)
    plt.xlabel("d"); plt.ylabel("log(Frequency)")
    plt.title("Antivaccine degree distribution log-histogram")
    plt.savefig("deg_mentions_before.eps"); plt.show()
    
    b = [i[1] for i in s2.degree()]
    b2 = collections.Counter(b)
    plt.bar(b2.keys(),b2.values())
    plt.xlabel("d"); plt.ylabel("Frequency")
    plt.title("Provaccine degree histogram after threshold value")
    plt.savefig("deg_mentions_after.eps"); plt.show()

def compare_cc(s1,s2):
    a = [round(i,2) for i in nx.clustering(s1).values()]   #Round to 0.01,0.02,etc.
    a2 = collections.Counter(a)
    a2v = list(a2.values())
    plt.bar(a2.keys(), np.log(a2v),width=0.01)
    plt.xlabel("Clustering Coefficient"); plt.ylabel("log(Frequency)")
    plt.title("Antivaccine clustering coefficient log-histogram")
    plt.savefig("cc_mentions_before.eps"); plt.show()
    
    b = [round(i,2) for i in nx.clustering(s2).values()]
    b2 = collections.Counter(b)
    plt.bar(b2.keys(),b2.values(),width=0.01)
    plt.xlabel("Clustering Coefficient"); plt.ylabel("Frequency")
    plt.title("Antivax Clustering Coefficient and Frequency in Mentions Network")
    plt.savefig("cc_mentions_anti.eps"); plt.show()
```

```python
compare_degree(heh1,heh2)
```

```python
compare_cc(heh1,heh2)
```

```python
def fraction_sent(s1,s2):      #Alter for other graphs
    c1 = list(nx.get_node_attributes(s1,"sent").values())
    print("P:",c1.count("P")/len(c1))
    print("N:",c1.count("N")/len(c1))
    d1 = list(nx.get_node_attributes(s2,"sent").values())
    print("P:",d1.count("P")/len(d1))
    print("N:",d1.count("N")/len(d1))
```

```python
fraction_sent(heh1,heh2)
```
